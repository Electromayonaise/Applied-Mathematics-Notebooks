## 1.  Use el principio de Cavalieri para plantear integrales que nos permitan calcular el volumen del sólido de revolución que se obtiene al girar la región plana acotada por las gráficas de las curvas $y=1-x^2$, el eje horizontal en torno a las rectas: a) $y=0$, b) $y=-1$,  c) $y=1$.


El principio de Cavalieri nos dice que el volumen de un sólido de revolución se puede calcular mediante la fórmula:


$$V = \pi \int_a^b [f(x)]^2 \, dx$$



Ahora, veamos cómo se derivan las integrales específicas para los diferentes ejes de rotación.

### 1.1 Rotación alrededor de $y = 0$ (Eje X)

Cuando la curva $y = f(x) = 1 - x^2$ se rota alrededor del eje $y = 0$, el volumen se obtiene utilizando la fórmula estándar para un sólido de revolución:

$$V = \pi \int_{-1}^{1} [f(x)]^2 \, dx = \pi \int_{-1}^{1} (1 - x^2)^2 \, dx$$

Aquí, estamos rotando la curva directamente alrededor del eje $y = 0$, lo que significa que simplemente tomamos el cuadrado de $f(x)$ y lo integramos sobre el rango de $x $ desde $-1$ hasta $1$.

### 1.2 Rotación alrededor de $y = -1$

Para rotar la curva alrededor de la recta $y = -1$, primero debemos ajustar la función para tener en cuenta este desplazamiento vertical. El radio de rotación en cualquier punto $x$ sería la distancia desde la curva hasta la línea $y = -1 $, que es $(1 - x^2) - (-1) = 1 - x^2 + 1 = 2 - x^2$.

El volumen se obtiene utilizando la fórmula:


$$V = \pi \int_{-1}^{1} \left( \text{Radio exterior}^2 - \text{Radio interior}^2 \right) \, dx $$

En este caso, el radio exterior es $ 2 - x^2$ y el radio interior es la distancia desde la recta de rotación hasta la línea $ y = -1 $, que es constante y vale 1:

$$V = \pi \int_{-1}^{1} \left((2 - x^2)^2 - 1^2\right) \, dx$$

### 1.3 Rotación alrededor de  $y = 1$

Para rotar la curva alrededor de la recta $ y = 1 $, nuevamente ajustamos la función. Aquí, la distancia desde la curva hasta $ y = 1 $ es 0.

El volumen se obtiene mediante:

$$ V = \pi \int_{-1}^{1} \left(\text{Radio exterior}^2 - \text{Radio interior}^2 \right) \, dx$$

En este caso, el radio exterior es $1$ (la distancia constante desde $ y = 1$ hasta el eje de rotación), y el radio interior es $1 - x^2$:

$$ V = \pi \int_{-1}^{1} \left(1^2 - (1 - x^2)^2\right) \, dx$$

## 2. Calcule analíticamente el volumen del sólido de la parte a) del problema anterior.


El volumen $V$ del sólido de revolución generado por la rotación de la curva $y = 1 - x^2$ alrededor del eje $y = 0$ se calcula mediante la integral:

$$V = \pi \int_{-1}^{1} (1 - x^2)^2 \, dx$$

Expandimos la función dentro de la integral:

$$(1 - x^2)^2 = 1 - 2x^2 + x^4$$

La integral se convierte en:

$$V = \pi \int_{-1}^{1} (1 - 2x^2 + x^4) \, dx$$

Integramos término a término:


$$\int_{-1}^{1} 1 \, dx = \left[ x \right]_{-1}^{1} = 1 - (-1) = 2$$

$$ \int_{-1}^{1} -2x^2 \, dx = -2 \int_{-1}^{1} x^2 \, dx = -2 \left[ \frac{x^3}{3} \right]_{-1}^{1} = -2 \left( \frac{1}{3} - \frac{-1}{3} \right) = -2 \left( \frac{2}{3} \right) = -\frac{4}{3}$$


$$\int_{-1}^{1} x^4 \, dx = \left[ \frac{x^5}{5} \right]_{-1}^{1} = \frac{1^5}{5} - \frac{(-1)^5}{5} = \frac{1}{5} - \frac{-1}{5} = \frac{2}{5} $$

Sumamos los resultados:

$$V = \pi \left( 2 - \frac{4}{3} + \frac{2}{5} \right)$$

Simplificando:

$$ V = \pi \left( \frac{30}{15} - \frac{20}{15} + \frac{6}{15} \right) = \pi \left( \frac{16}{15} \right)$$

Por lo tanto, el volumen del sólido es:

$$V = \frac{16\pi}{15} ≈ 3.351$$



## 3. Use las rutinas implementadas para calcular numericamente los volumenes del ejercicio 1, con $n=100$.

```python
# 3.1

import numpy as np

def f(x):
    return 1 - x**2

a, b = -1, 1
n = 100
delta_x = (b - a) / n
total_sum = 0
for i in range(n):
    mid_point = a + (i + 0.5) * delta_x
    total_sum += f(mid_point)**2

volume = total_sum * delta_x * np.pi
print(volume)
```

## Analítico

Para calcular el volumen del granero usando integrales dobles, necesitamos establecer la integral sobre la base rectangular del granero e integrar la función altura.

- Dimensiones de la base: 20 pies (anchura a lo largo del eje y) por 40 pies (longitud a lo largo del eje x).
- Altura en la parte delantera (x = 40): 30 pies.
- Altura trasera (x = 0): 40 pies.

La altura del granero $z(x)$ cambia linealmente de 30 pies en la parte delantera a 40 pies en la parte trasera. Esto se puede representar como:
$$ z(x) = 40 - \left(\frac{10}{40}\right)x = 40 - 0.25x $$

Para encontrar el volumen, integramos la función de altura sobre el área de la base:
$$ V = \int_{0}^{20} \int_{0}^{40} h(y) \, dx \, dy $$

Sustituyendo:
$$ V = \int_{0}^{20} \int_{0}^{40} (40 - 0,25x) \, dx \, dy $$

En primer lugar, evaluar la integral con respecto a \( y \):
$$ \int_{0}^{40} (40x - 0,125x^2) \, dy = \left[ 40x - 0,125x^2 \right]_{0}^{40} = 40(40) - 0,125(40^2) = 1600 - 200 = 1400 $$

Ahora, integrar con respecto a \( x \):
$$ V = \int_{0}^{20} 1400 \, dx = 1550y \Big|_0^{20} = 1400(20) = 28000 \, \text{pies cúbicos} $$

Por tanto, el volumen del granero es **28.000 pies cúbicos**.

```python
# Númericamente

def f(x,y):
  return (-x/4)+40

a,b,c,d= 0,40,0,20
n=1000
m=1000
dx=(b-a)/n
dy=(d-c)/m
sum=0
for j in range(m):
  for i in range(n):
    x=a+i*dx
    y=c+j*dy
    sum+=f(x,y)*dx*dy
print(sum)
```
