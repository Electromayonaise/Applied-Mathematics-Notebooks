#Integrales dobles sobre regiones mas generales

Supongamos que se tienen dos funciones continuas de valor real $φ_1 : [a, b] → \mathbb{R}$ y $φ_2: [a, b] → \mathbb{R}$ que satisfacen $φ_1(x) ≤ φ_2(x)$ para todo $x ∈ [a, b]$. Sea $D$ el conjunto

$$D=\{(x,y): x ∈ [a, b], φ_1(x) ≤ y ≤ φ_2(x)\}.$$

Esta región $D$ se dice es y-simple. La figura siguiente muestra varios ejemplos de regiones $y-$simples.

# Definición (Integral sobre una región simple)
Si $D$ es una región simple en el plano, $R$ un rectángulo que contiene a $D$ y  $f : D → \mathbb{R}$, donde $f$ es continua ( y por lo tanto acotada ), entonces la integral doble sobre la región $D$,

$$\int\int_D f (x, y) \,dA$$

se define haciendo uso de una función $f^{\ast}$ sobre el rectángulo $R$, como sigue: Extienda $f$ a una función $f^∗$ definida en todo $R$ por

$$f^*(x, y)=\begin{cases}
  f (x, y), & \text{si }(x, y) ∈ D  \\
  0 & \text{ si } (x, y) \not∈ D,\,(x, y)∈ R
\end{cases}$$

Por lo tanto, podemos definir la integral doble sobre la región $D$, de la forma

$$\int\int_D f (x, y)\, dA =\int\int_R f^∗(x, y)\, dA.$$




Es fácil mostrar la validez de los siguientes resultados.


#Teorema (Reducción a integrales iteradas)
Si $D$ es una región $y-$simple entonces
$$\int\int_D f (x, y)\, dA =\int_a^b\int_{φ_1(x)}^{φ_2(x)}f (x, y)\, dy dx.$$
En el caso que $D$ sea una región $x-$simple entonces
$$\int\int_D f (x, y)\, dA =\int_c^d\int_{φ_1(y)}^{φ_2(y)}f (x, y)\, dx dy.$$

#Cambio en el orden de integración

Suponga que $D$ es una región simple, es decir, es tanto $x-$simple como $y-$simple. Por lo tanto, se puede dar como el conjunto de puntos $(x, y)$ tal que
$$a ≤ x ≤ b, φ_1(x) ≤ y ≤ φ_2(x),$$
y también como el conjunto de puntos $(x, y)$ tal que
$$c ≤ y ≤ d, ψ_1( y) ≤ x ≤ ψ_2( y).$$

Por lo tanto, tenemos las fórmulas
$$\int\int_D f (x, y)\, dA =
\int_a^b \int_{φ_1(x)}^{φ_2(x)} f (x, y)\, dy dx =\int_c^d \int_{ψ_1(y)}^{ψ_2(y)}f (x, y)\,dx dy.$$

Si tenemos que calcular una de las integrales iteradas anteriores, podemos hacerlo evaluando la otra integral iterada; esta técnica se llama cambiar el orden de integración. Puede ser útil hacer tal cambio al evaluar integrales iteradas, porque una de las integrales iteradas puede ser más difícil de calcular que la otra.
