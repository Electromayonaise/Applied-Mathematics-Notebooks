La forma general de una ecuación diferencial de primer orden se presenta como $\frac{dy}{dx} + P(x)y = Q(x)$, en donde $P(x)$ y $Q(x)$ representan funciones que dependen de $x$. Para abordar la resolución de este tipo de ecuaciones, podemos recurrir al método del factor integrante. Este método implica la multiplicación de ambos lados de la ecuación por un factor denominado factor integrante, seguido de la integración de ambos lados.

El método del factor integrante es una herramienta esencial para resolver ecuaciones diferenciales lineales de primer orden. Su estrategia radica en seleccionar un factor integrante que permita transformar la ecuación resultante en una forma más manejable.

La ecuación diferencial lineal de primer orden, en su formato genérico $dy/dx + P(x)y = Q(x)$, se conecta con el factor integrante mediante la fórmula $e^{\int P(x)dx}$, donde $\int P(x)dx$ representa la integral indefinida de $P(x)$ con respecto a $x$.

Una vez calculado el factor integrante, se procede a multiplicar ambos lados de la ecuación por este. La ecuación resultante se configura como $e^{\int P(x)dx}\frac{dy}{dx} + P(x)y = e^{\int P(x)dx}Q(x)$. Mediante la regla del producto, esta ecuación puede ser expresada como $\frac{d}{dx}[e^{\int P(x)dx}y] = e^{\int P(x)dx} Q(x)$. Con este punto de partida, la integración de ambos lados de la ecuación nos conduce hacia la obtención de la solución general.




\begin{alignat*}{2}
\text{1.} \quad & \frac{dy}{dx} + P(x)y = Q(x) \quad && \text{Ecuación diferencial dada} \\
\text{2.} \quad & e^{\int P(x) \, dx} \cdot \frac{dy}{dx} + P(x)e^{\int P(x) \, dx}y = e^{\int P(x) \, dx}Q(x) \quad && \text{Multiplicamos por factor integrante} \\
\text{3.} \quad & \frac{d}{dx} \left( e^{\int P(x) \, dx}y \right) = e^{\int P(x) \, dx} \cdot \frac{dy}{dx} + P(x)e^{\int P(x) \, dx}y \quad && \text{Derivada del producto} \\
\text{4.} \quad & \frac{d}{dx} \left( e^{\int P(x) \, dx}y \right) = e^{\int P(x) \, dx}Q(x) \quad && \text{Sustituimos con } Q(x) \\
\text{5.} \quad & \int \frac{d}{dx} \left( e^{\int P(x) \, dx}y \right) \, dx = \int e^{\int P(x) \, dx}Q(x) \, dx \quad && \text{Integramos ambos lados} \\
\text{6.} \quad & e^{\int P(x) \, dx}y = \int e^{\int P(x) \, dx}Q(x) \, dx  \quad && \text{Resolvemos la integral en el lado derecho} \\
\text{7.} \quad & y = \frac{\int e^{\int P(x) \, dx}Q(x) \, dx}{e^{\int P(x) \, dx}} \quad && \text{Dividimos por el factor integrante} \\
\end{alignat*}







Consideremos la siguiente ecuación diferencial: $$\frac{dy}{dx} + \frac{3y}{x} = \frac{1}{x^2}$$

Paso 1: Identificación

Identificamos \(P(x)\) y \(Q(x)\):
$$P(x) = \frac{3}{x}, \quad Q(x) = \frac{1}{x^2}$$

Paso 2: Factor Integrante

Encontramos el factor integrante $$e^{\int P(x) \, dx}$$
$${\int P(x) \, dx} =\int \frac{3}{x} \, dx = 3\ln|x| + C$$
Factor integrante: $$e^{\int P(x) \, dx} = e^{3\ln|x| + C} = c_1x^3$$
Paso 3: Solución General

La solución general es:
$$
y = \frac{\int c_1x^3 \cdot \frac{1}{x^2} \, dx}{c_1x^3} = \frac{c_1 \int x \, dx}{c_1x^3}
$$

$$
= \frac{ \frac{x^2}{2} + K}{x^3} = \frac{1}{2x} + \frac{K}{x^3}
$$

Donde \(K\) es una constante de integración.

Hemos resuelto la ecuación diferencial utilizando el factor integrante. La solución general es $$y = \frac{1}{2x} + \frac{K}{x^3}$$







#Metodo para resolver ecuaciones diferenciales lineales de primer orden#

```python
from sympy import symbols, Function, Eq, dsolve

# Definir las variables simbólicas
x = symbols('x')
y = Function('y')(x)

# Definir la ecuación diferencial
equation = Eq(y.diff(x) + 3*y/x, 1/x**2)

# Resolver la ecuación diferencial
solution = dsolve(equation, y)

print(solution)
```

#Metodo para encontrar la solucion general y valor numerico para una ecuacion diferencial con condiciones iniciales.#

```python
import sympy as sp

# Define las variables simbólicas
x = sp.symbols('x')
y = sp.Function('y')(x)

# Define la ecuación diferencial
ecuacion_diff = sp.Eq(y.diff(x), 6-((3/(40-x))*y))

# Define las condiciones iniciales
condiciones_iniciales = {y.subs(x, 0): 1}

# Resuelve la ecuación diferencial con las condiciones iniciales
solucion = sp.dsolve(ecuacion_diff, y, ics=condiciones_iniciales)
print(solucion)

# Define la solución general
solucion_general = solucion.rhs

# Evalúa la solución en x = 1.8
valor_y_1_8 = solucion_general.subs(x, 1.8).evalf()

# Muestra los resultados
print("Solución general:", solucion_general)
print("y(1.8) =", valor_y_1_8)
```

#¿Qué velocidad inicial $v_{0}$ es necesaria para que un proyectil de masa $m$ escape por completo de la Tierra?

Partimos de que $\forall_{t>0}\:\frac{dr}{dt} > 0$ y que $r(t)$ representa la distancia del proyectil desde el centro de la Tierra en el tiempo $t$.

Y consideramos que cuando $r→∞$ entonce $v→0$

Sabemos que $\frac{dv}{dt} = \frac{d^2r}{dt^2} = -\frac{GM}{r^2}$, esto es debio a que la aceleración que sufre el cuerpo es la que ejerce la tierra sobre el objeto.

Haciendo un análisis podemos observar esto de mejor manera:

Tenemos la segunda ley de Newton: $\sum_{i=1}^{n} F_{i} = ma$, ahora usando la ley de gravitación universal tenemos: $\sum_{i=1}^{n} F_{i} = \frac{-GmM}{r^2}$

---
Igualamos y obtenemos que: $\frac{-GmM}{r^2} = ma$, despejando $a$ obtenemos:
$$a = \frac{-GM}{r^2}$$ Que es lo mismo que decir: $\frac{dv}{dt} = \frac{d^2r}{dt^2} = \frac{-GM}{r^2}$



Tenemos: $$v = \frac{dr}{dt} \wedge \frac{d^2r}{dt^2} = \frac{dv}{dt}$$

Por lo que podemos operar de forma algebraica $\frac{dv}{dt}$ mulplicandolo por $\frac{dr}{dr}$ y no altera nuestra expresión debido a nuestra aceleración está en función de $r$

$$\frac{dv}{dt} = \frac{dv}{dt}\frac{dr}{dr}\\
\frac{dv}{dt} = \frac{dv}{dr}\frac{dr}{dt}$$

Observamos que $\frac{dr}{dt} = v$, por lo que nos queda que $\frac{dv}{dt} = v\frac{dv}{dr}$

Lo que a su vez implica que $v\frac{dv}{dr} = -\frac{GM}{r^2}$

Ahora integramos ambos lados respecto a r y nos queda:

$$\int v \frac{dv}{dr} dr = \int -\frac{GM}{r^2}dr\\
\int v dv = -GM \int \frac{1}{r^2}dr\\
\frac{v^2}{2} = \frac{GM}{r} + C$$

Sabemos que $t=0$ implica que: $v = V_{o}$ y $r = R$, debido a que
es el punto de salida. Con esto podemos obtener $C$

$$C = \frac{v_{o}^2}{2} - \frac{GM}{R}$$

Reemplazamos $C$ en nuestra ecuación y nos queda:

$$\frac{v^2}{2} = \frac{GM}{r} + \frac{v_{o}^2}{2} - \frac{GM}{R}$$

$$v^2 = 2\frac{GM}{r} + v_{o}^2 - 2\frac{GM}{R}$$

$$v^2 = v_{o}^2 + 2GM(\frac{1}{r} - \frac{1}{R})$$

Ahora debido a que cuando $r→∞$ entonces $v→0$, calculamos dichos límites en la ecuación obtenida:

$$\lim_{v\rightarrow 0} v^2 = \lim_{r\rightarrow \infty} v_{o}^2 + 2GM(\frac{1}{r}-\frac{1}{R})\\
0 = v_{o}^2 - \frac{2GM}{R}\\
v_{o}^2 = \frac{2GM}{R}
$$

Por lo tanto la $V_{e}$ es:

$$V_{e} = \sqrt{\frac{2GM}{R}}$$

#Trayectorias de vuelo#

Las trayectorias de vuelo en el contexto de ecuaciones diferenciales se refieren a las curvas o caminos que describe un objeto, como un avión o un proyectil, en función del tiempo u otra variable independiente. Estas trayectorias se obtienen a partir de ecuaciones diferenciales que relacionan las tasas de cambio de las variables de posición del objeto con respecto al tiempo o a otra variable independiente.

En términos más simples, cuando un objeto se mueve en el espacio y su movimiento está influenciado por fuerzas o condiciones específicas, podemos modelar ese movimiento utilizando ecuaciones diferenciales. Estas ecuaciones describen cómo cambian las posiciones del objeto a medida que transcurre el tiempo o alguna otra variable. La solución de estas ecuaciones diferenciales proporciona las ecuaciones paramétricas o funcionales que definen la trayectoria que el objeto sigue a lo largo de su movimiento.

Por ejemplo, en el caso de un avión que vuela en un entorno con viento constante, las ecuaciones diferenciales pueden modelar cómo cambian las coordenadas x e y del avión con respecto al tiempo. Estas ecuaciones pueden tener en cuenta tanto la velocidad del avión en relación con el viento como la velocidad del viento en sí. La solución de estas ecuaciones diferenciales resulta en las ecuaciones de la trayectoria del avión en el espacio bidimensional.

El uso de ecuaciones diferenciales para modelar trayectorias de vuelo es crucial en la aviación, la navegación y la ingeniería aeroespacial. Permite predecir y comprender cómo un objeto se moverá en función de diferentes condiciones y parámetros, lo que es esencial para la planificación de rutas, la navegación precisa y el diseño de vehículos aéreos.

El planteamiento de un modelo general resulta complicado, pues depende de muchos factores, cómo la dirección a la que viaja el objeto y que fuerzas actúan sobre el mismo. Además al tratar con este tipo de ecuaciones, por ejemplo, solo podemos tratar con 2 fuerzas que actúen sobre el objeto, pues el usar más implicaría más diferernciales. ## Código que representa la trayectoria de vuelo

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_trajectory(a, v_0, w):
    k = w / v_0
    x = np.linspace(-a, a, 1000)
    y = a/2 * ((x/a)**(1-k) - (x/a)**(1+k))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y(x)')
    plt.title('Trajectory of a Plane')
    plt.show()

# Example usage:
plot_trajectory(a=200, v_0=500, w=100)
```

***Casos de fuerzas no constantes***

Consideremos un caso en el que se aplica una fuerza paralela al desplazamiento. Ademas, dicha fuerza no es constante y se pueda expresar como una función de la posición x.

work3.png

En ese caso, una aproximación podría consistir en asumir que la fuerza fue constante en pequeños intervalos Δx, para luego sacar el trabajo producido en cada uno estos y sumarlos.


$$W =\lim_{{\Delta x \to 0}}
 \sum_{i=1}^{n} F_i \cdot \Delta x$$




Esto vendría a hacer una integral o suma de riemann.

$$W = \int_{x_i}^{x_f} f(x) \cdot dx$$

De acuerdo al sistema internacional, el trabajo se mide N*M. Lo que equivale a kg *m/s

***Trabajo neto y teorema de trabajo energía***

Si consideraramos el trabajo producido, no por una fuerza en específico, sino por la sumatoria de fuerzas, ya se estaría calculando el trabajo neto. La expresión pasaría a tener esta forma.
$$W_{\text{neto}} = \int_{x0}^{xf} \sum F \cdot dx$$

Por segunda ley de newton se sabe que
$$\sum F= m*a$$
Teniendo presente que en el tipo de problemas que se va a analizar la masa es una constante y la aceleración es una la tasa de variación instantanea de la velocidad con respecto al tiempo se podría realizar la siguiente sustitución
$$\sum F= m*\frac{dv}{dt}$$

Sustituyendo en la ecuación de trabajo neto se tiene que:
$$W_{\text{neto}} = \int_{x0}^{xf}( m*\frac{dv}{dt}) \cdot dx$$
Dado que la masa es una constante se puede sacar de la integral.
$$W_{\text{neto}} = m\int_{x0}^{xf} \frac{dv}{dt} \cdot dx$$
Reescribiendo la expresión (intercambiando el dv y el dx)

$$W_{\text{neto}} = m\int_{x_{0}}^{x_{f}}( \frac{dx}{dt}) \cdot dv$$
Si se tiene encuenta que $v=\frac{dx}{dt}$.
Se podría expresar toda la integral en terminos de la función v.
$$W_{\text{neto}} = m\int_{vi}^{vf}( v \cdot dv)$$
Se integra usando la regla de la potencia y teniendo en cuenta que se está calculando una integral definida
$$W_{\text{neto}} = m(  \frac {vf^2}{2}-\frac {vi^2}{2} )$$
$$W_{\text{neto}} =  m\frac {vf^2}{2}-m\frac {vi^2}{2} $$

Este resultado hace que el calculo del trabajo se simplifique mucho y se pueda ver como una simple diferencia. Tambien se creó un concepto llamado energía cinetica, representado como $K=\frac{1}{2}mv^2$.
$$W_{\text{neto}} =  K_f-K_i=ΔK$$

Tambien se podría hacer uso de metodos como el de simpson 1/3 y 3/8 para halllar este tipo de integrales

```python
import sympy as sp
import numpy as np
def f(x):
  return (sp.cos(x+1))+x**2;
def simpson13(x_0,x_1,x_2):
  h=x_1-x_0
  return ((h)*(1/3))* (f(x_0)+4*f(x_1)+f(x_2))
def simpson38(x_0,x_1,x_2,x_3):
  h=x_1-x_0
  return (h*3/8)*(f(x_0)+3*f(x_1)+3*f(x_2)+f(x_3))

a=1 #extremo inferior
b=4 #extremo superior
n=2001 #Numero de intervalos
h= abs( (b-a))/n            #distancia entre x_i y x_i+1
puntos=np.arange(a,b+h,h)
ans=0.0

if n%2==0:

 for i in range(0,len(puntos)-2,2):# si n es par simplemente aplico simpson 1/3 para todos
  ans+=simpson13(puntos[i],puntos[i+1],puntos[i+2])


else:
  for i in range(0,len(puntos)-4,2): #si n es impar aplico simpson 1/3 para casi todos, solo dejo que me sobren 4 puntos
    ans+=simpson13(puntos[i],puntos[i+1],puntos[i+2])
  ans+=simpson38(puntos[len(puntos)-4],puntos[len(puntos)-3],puntos[len(puntos)-2],puntos[len(puntos)-1]) # a esos 4 ultimos sobrantes se les aplica simpson 3/8

print(ans)
```

#Uso de la densidad lineal de una varilla#

Se tiene una barra de longitud \(L\) colocada a lo largo de un eje \(x\), cuya densidad lineal varía en función de la posición. La densidad en un punto \(x\) está representada por la función \(f(x)\). El objetivo es calcular la masa total de la barra a partir de esta distribución de densidad.


Para determinar la masa total de la barra, se utiliza el concepto de integral definida. La masa \(m\) de una pequeña porción de la barra entre dos puntos \(a\) y \(b\) en el eje \(x\) se expresa como \(R(f, [a,b])\), donde \(p(x)\) es la densidad lineal en el punto \(x\). Si la densidad es constante, la masa se calcula como $$p(x) \cdot (b - a)$$

Si la densidad \(f(x)\) es variable, se establece un rango de densidades posibles entre \(m\) y \(M\), donde \(m\) es la densidad mínima y \(M\) es la densidad máxima de la barra. Luego, se demuestra que $$m(b - a) \leq R(f, [a,b]) \leq M(b - a)$$

Para realizar una aproximación numérica de la masa, se divide el intervalo \([a, b]\) en \(n\) segmentos de igual longitud $$\Delta x = \frac{b - a}{n}$$. Cuanto mayor sea \(n\), más precisa será la aproximación. La masa total de la barra se obtiene sumando las masas de cada segmento calculadas con las densidades \(f(c_i)\) en puntos \(c_i\) dentro de cada subintervalo \([a, c_i]\).

Mediante este enfoque, se logra calcular la masa de una barra con densidad variable utilizando conceptos de cálculo integral y aproximación numérica.

En caso de que la densidad de la varilla sea constante podemos expresar la masa de la varilla como una simple multiplicacion, puesto que si multiplicamos la densidad * la longitud nos daria la masa $$ m = p(x) * (b-a)$$

Sin embargo, cuando la densidad no es constante podemos segmentar la varilla en deltas $$\Delta x$$ para poder encontrar la mi siendo una segmento de la masa tal que $$mi = p(xi)(x_i- x_{i-1}) = p(x)*\Delta x$$

Haciendo la sumatoria de los segmentos

$$M = \sum_{i=1}^{n} mi$$
$$M = \sum_{i=1}^{n} p(x)*\Delta x$$

Utilizando el limite de n -> $\infty$ encontramos la formula que nos define la integral para hallar la masa de varilla.

$$M = \lim_{{n \to \infty}} \sum_{i=1}^{n} p(x)*\Delta x$$
$$M = \int_{a}^{b} p(x) \, dx$$


Explicado con la sumatoria de las partes de la varilla:

$$m\leq p(x)\leq M$$

siendo p(x) la densidad para una parte de la masa, se entinede que la minima densidad lineal es menor o igual que p(x) y p(x) sera menor o igual que la maxima densidad lineal puesto a que este entre este intervalo.

siendo R(f, [a,b]) la masa de [a,b] y $\Delta x$ es la longitud de cada partición $[a,c_i]$ siendo $$c_i \in [a,b]$$

Podemos definir la suma total de la masa de la varilla de tal forma que:
$$R(f, [a,b]) = R(f, [a,c_1]) +...+ R(f, [c_n,b])$$

Ejercicio 1: Densidad Constante

Supongamos que tenemos una barra de longitud \(L = 8\) metros con densidad constante $$f(x) = 3 \, \text{kg/m}$$ Queremos calcular la masa total de la barra.

La masa \(m\) se calcula utilizando la fórmula \(R(f, [a,b]) = f(x)(b - a)\):

$$ m = R(f, [0, 8]) = f(x)(b - a) = 3 \, \text{kg/m} \times 8 \, \text{m} = 24 \, \text{kg} $$


Ejercicio 2: Densidad Variable y Suma de Riemann

Supongamos que tenemos una barra de longitud \(L = 10\) metros con densidad variable:

\[ f(x) = \begin{cases}
    2x & \text{si } 0 \leq x \leq 5 \\
    4x & \text{si } 5 < x \leq 10
\end{cases}\]

Queremos calcular la masa total de la barra utilizando la integral definida y la aproximación de la suma de Riemann.

Usando la Integral Definida

La masa \(m\) se calcula integrando la densidad variable \(f(x)\) en el intervalo \([0, 10]\):

$$ m = \int_{0}^{10} f(x) \, dx = \int_{0}^{5} 2x \, dx + \int_{5}^{10} 4x \, dx $$

Calculamos las integrales:

$$ m = \left. x^2 \right|_{0}^{5} + \left. 2x^2 \right|_{5}^{10} = 25 + 200 - 0 - 50 = 175 \, \text{kg} $$





```python
# Ejercicio 1: Densidad Constante

def ejercicio_densidad_constante(longitud, densidad):
    masa = densidad * longitud
    return masa

longitud1 = 8  # metros
densidad1 = 3  # kg/m

masa_ejercicio1 = ejercicio_densidad_constante(longitud1, densidad1)
print(f"Ejercicio 1: Masa total = {masa_ejercicio1} kg")

import sympy as sp

def ejercicio_densidad_variable(densidad_funcion, intervalo):
    x = sp.symbols('x')
    masa = sp.integrate(densidad_funcion, (x, intervalo[0], intervalo[1]))
    return masa

x = sp.symbols('x')

densidad_funcion = sp.Piecewise((2 * x, (x >= 0) & (x <= 5)), (4 * x, (x > 5) & (x <= 10)))
intervalo_ejercicio2 = (0, 10)

masa_ejercicio2 = ejercicio_densidad_variable(densidad_funcion, intervalo_ejercicio2)
print(f"Ejercicio 2: Masa total = {masa_ejercicio2} kg")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de densidad variable como Piecewise
def densidad_variable(x):
    if 0 <= x <= 5:
        return 2 * x
    elif 5 < x <= 10:
        return 4 * x
    else:
        return 0  # Valor fuera de los intervalos

# Creación de valores de x para graficar
x_vals = np.linspace(0, 10, 400)

# Evaluación de la función de densidad en los valores de x
y_vals = [densidad_variable(x) for x in x_vals]

# Creación del gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$f(x)$')
plt.xlabel('x')
plt.ylabel('Densidad $f(x)$')
plt.title('Función de Densidad Variable')

# Área bajo la curva
a = 0
b = 10
x_fill = np.linspace(a, b, 100)
y_fill = [densidad_variable(x) for x in x_fill]
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.3, label='Área bajo la curva')

# Cálculo del área utilizando el método de los rectángulos
n_rectangulos = 1000
delta_x = (b - a) / n_rectangulos
area_aproximada = 0
for i in range(n_rectangulos):
    x_i = a + i * delta_x
    area_aproximada += densidad_variable(x_i) * delta_x

plt.legend()
plt.grid(True)
plt.annotate(f"Área(Masa barra) ≈ {area_aproximada:.2f}", xy=(2, 6), fontsize=12)
plt.show()
```
