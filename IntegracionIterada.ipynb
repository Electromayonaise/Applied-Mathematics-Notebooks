Esta manera es trivial, pues basta con hallar el área del tríangulo tomandolo como base, y luego cultiplicarlo por la altura que sería z. Esto nos da $\frac{5⋅3}{2}⋅4 = 30$

## 2.  

Usando integrales dobles primero planteados


$$\int \int_R (3-\frac{3}{5}x)dA$$
$$\int_0^4 \int_0^5 (3-\frac{3}{5}x)dxdy$$
$$\int_0^4  (3x-\frac{3}{5}\cdot \frac{x^2}{2})\Big|_0^5dy$$
$$\int_0^4  (15-\frac{15}{2})dy$$
$$\int_0^4  (\frac{15}{2})dy$$
$$ \frac{15}{2}y\Big|_0^4 = 30$$

## 3.

```python
def f(x,y):
  return 3-(3/5)*x

a,b,c,d= 0,5,0,4
n=1000
m=1000
dx=(b-a)/n
dy=(d-c)/m
sum=0
for j in range(m):
  for i in range(n):
    x=a+i*dx
    y=c+j*dy
    sum+=f(x,y)*dx*dy
print(sum)
```

## *Ejercicio 2*

Plantee una integral doble para encontrar el volumen de un cilindro circular recto de radio de la base $r$ y altura $h$.

Primero imaginemonos que cortamos el cilindro a la mitad, de manera que obtenemos un

$$ R= [-r,r] \times [0,h]$$

Ahora, el semicirculo de la base se define como

$$ x^2 + z^2 = r^2$$

por lo que

$$ z = \sqrt(r^2-x^2)  = f(x,y)$$

Por lo tanto, la integral doble que me da el volumen es

$$ V = 2\int_{-r}^r \int_0^h\sqrt(r^2-x^2)dydx$$

(Se multiplica por 2 dado que al inicio lo cortamos a la mitad)

Ahora dado que sabemos que esto debe ser igual a el área del círculo base por la altura del cilindro

$$ V = 2\int_{-r}^r \int_0^h\sqrt(r^2-x^2)dydx = πr^2h$$

Podemos confirmarlo en python

```python
def f(x,y):
  return (2*(1-x**2)**0.5)

a,b,c,d= -1,1,0,1
n=10000
m=1000
dx=(b-a)/n
dy=(d-c)/m
sum=0
for j in range(m):
  for i in range(n):
    x=a+i*dx
    y=c+j*dy
    sum+=f(x,y)*dx*dy
print(sum)
```

Cuando tenemos un sólido, sea su base rectangular o no, siempre podemos calcular el vólumen de sus secciones transversales como

$v = A(x)Δx$ y el vólumen del sólido será $∑v = V$

Es decir, en caso de que el área de integración sea un rectángulo, tenemos entonces que


$$V ≈ \sum_{i=1}^n A(x_i)Δx$$

$$ V = \int_a^b A(x)dx = \int_a^b\int_c^df(x,y)dydx$$

Esto aplica igualmente en el ejercicio pasado, pues es un sólido de revolución, que se puede obtener al girar un rectangulo.

Para un sólido de revolución, todas las secciones transversales serán discos
