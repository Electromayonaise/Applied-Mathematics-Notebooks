#Cuadricas
Las superficies cuádricas son los gráficos de las ecuaciones que, en un plano tridimensional, pueden expresarse en la forma

$$Ax^2 + By^2 + Cz^2 + Dxy + Exz + Fyz + Gx + Hy + Jz + K = 0$$

Existen seis tipos básicos de superficies cuádricas: Elipsoide, Hiperboloide de una hoja, Hiperboloide
de dos hojas, Cono Elíptico, Paraboloide Elíptico y Paraboloide Hiperbólico.

#Trazas
Estas consisten en graficas que surgen cuando se intersecta un objeto en tres dimensiones con un plano de dos dimensiones.

Por lo general, los planos bidimensionales utilizados son paralelos a xy, xz o yz.

La forma en la que se calcula la ecuación de una traza consiste en tomar la ecuación de la cuadrica e igualar a 0 alguna de las variables presentes, dependiendo del plano de interes.

#Elipsoide
Su ecuación canónica es:
$\frac {X^2}{a^2}+\frac {y^2}{b^2}+\frac {z^2}{c^2}=1$

a,b y c son los semiejes y representan la apertura del elipsoide en el eje "x", "y" y "z" respectivamente.

Tambien todas las trazas que se generan a partir del elipsoide y que pasan por el origen , son elipses.
$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$

#Hiperboloide de una hoja 

Ecuación Canónica

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1
$$

Si el plano es paralelo xy, la traza es una elipse. Si el plano es paralelo xz o yz se forman una hiperbola.

image.png

#Hiperboloide de 2 hojas
Su ecuación canónica es:

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$$

Al igual que el hiperboloide de 1 hoja, Si el plano es paralelo xy, la traza es una elipse y si el plano es paralelo xz o yz se forman una hiperbola.

image.png

#Cono Elíptico
Su ecuación canónica es:
$${\frac{x^2}{a^2} + \frac{y^2}{b^2} = z^2}$$

Si el plano es paralelo al eje xy, la traza es una elipse. Si es el xz o yz es una hiperbola

#Paraboloide elíptico
Su ecuación canónica es:

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = z$$

Las trazas resultantes pueder ser una elipse si el plano es paralelo al eje xy, y una parabola si es paralelo al xz o yz

##Cuádrica seleccionada: Paraboloide Hiperbólico

El paraboloide hiperbólico es una superficie engendrada por el desplazamiento de una parábola generatriz que se desliza paralelamente a sí misma a lo largo de otra parábola directriz de curvatura opuesta situada en su plano de simetría.

El paraboloide hiperbólico tiene la particularidad de ser, al igual que el hiperboloide de una hoja, una superficie reglada, contando con dos familias de rectas que podrían entenderse respectivamente como directrices y generatrices.

Matemáticamente, se puede expresar como:

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} = z $$



$$
\begin{align*}
& En\ esta\ ecuación: \\
&- x, y, \text{ y } z \text{ son las coordenadas tridimensionales de un punto en el espacio.} \\
&- a \text{ y } b \text{ son constantes que determinan la forma del paraboloide hiperbólico.} \\
&- a \text{ controla la apertura del paraboloide en el eje } x \text{ y } b \text{ controla la apertura en el eje } y.
\end{align*}
$$

###Secciones planas del paraboloide hiperbólico

Como se mencionó anteriormente, el paraboloide hiperbólico tiene la peculiaridad de que dependiendo de la posición del plano de corte las secciones del paraboloide pueden ser:
- Parábolas, si el plano de corte es paralelo al eje del paraboloide. Si además es paralelo a uno de los planos directores, la intersección degenera en una recta.
- Hipérbolas, si el plano secante corta al eje del paraboloide. Si lo corta ortogonalmente y pasa además por el centro de gravedad de la superficie, la intersección degenera en dos rectas.

### Código para generar un paraboloide hiperbólico

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a = 1.0
b = 1.0

# Crear una malla de valores para x y y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Calcular la ecuación del paraboloide hiperbólico
z = (x**2 / a**2) - (y**2 / b**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar el paraboloide hiperbólico
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.title('Paraboloide Hiperbólico')

# Mostrar la gráfica
plt.show()
```

Caso de Uso: Diseño de un Pabellón de Exposiciones

Supongamos que un arquitecto está diseñando un pabellón de exposiciones futurista. El objetivo es crear una estructura que no solo sea funcional, sino que también sirva como punto focal de interés y admiración en el entorno urbano. El arquitecto tiene en mente un diseño con una curvatura en el techo que siga la forma de un paraboloide hiperbólico. Para lograr esto, decide utilizar un paraboloide con una curvatura de 2 unidades y 1.5 orientación en la dirección del eje y.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

curvatura = float(input("Ingresa el valor de la curvatura en la direccion del eje x : ")) # Curvatura en la dirección del eje x
orientacion = float(input("Ingresa el valor de la orientación en la direccion del eje y: ")) # Orientación en la dirección del eje y

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = (x**2 / curvatura**2) - (y**2 / orientacion**2)
z = -z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.title('Paraboloide Hiperbólico (Estilo Silla de Montar)')

plt.show()
```
